```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("TeluguMovies_dataset.csv",skipinitialspace=True)
df.fillna("")
df1=df
df
```

```python
df.dtypes
```

```python
cols=["Movie","Certificate","Genre","Overview"]
for col in cols:
    df[col]=df[col].str.strip()
    df[col]=df[col].str.replace(',','')
    df[col]=df[col].fillna('')
```

```python
df
```

```python
def combine(row):
    try:
        return row["Genre"]+" "+row["Overview"]
    except:
        print("Error : ",row)
```

```python
df["all_aspects"]=df.apply(combine,axis=1)
df.fillna(" ")
```

```python
df["all_aspects"].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
count_matrix = cv.fit_transform(df["all_aspects"])
```

```python
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
co_sim=cosine_similarity(count_matrix)
```

```python
co_sim
```

```python
def movie_index(movie_name):
    return df[df.Movie == movie_name]["index"].values[0]
```

```python
movie_index("Jakkanna")
```

```python
def movie_title(movie_index1):
    return df[df.index == movie_index1]["Movie"].values[0]
```

```python
similar_movies=list(enumerate(co_sim[movie_index(input("Enter movie name : "))]))
```

```python
similar_movies=sorted(similar_movies,key=lambda x:x[1],reverse=True)
```
